```python
import numpy as np
from scipy import stats

comps =  [['231P','231T'],['231P','PBS'], ['PBS','5UTEVp'],['5UTEVp', '10UTEVp'],['10UTEVp', '20UTEVp']]
for one, two in comps:
    first=allpoints['FR_nuc_mem_conc_ratio'][allpoints['FR_nuc_mem_conc_ratio']['condition']==one]['value_mean']
    second=allpoints['FR_nuc_mem_conc_ratio'][allpoints['FR_nuc_mem_conc_ratio']['condition']==two]['value_mean']

    tstat, pval = stats.ttest_ind(first, second, equal_var=False)
    mean1, mean2 = first.mean(), second.mean()
    print(
    f"{one} vs {two}: "
    f"mean={mean1:.3f} vs {mean2:.3f}, "
    f"t = {tstat:.2f}, p = {pval:.3g}"
    )
```

```python
allpoints['norm_GFP_mem']

import numpy as np
from scipy import stats

comps = [['5UTEVp_early','5UTEVp_late'], ['PBS_early','PBS_late']]
for early, late in comps:
    before=allpoints['norm_GFP_mem'][allpoints['norm_GFP_mem']['cond_bin']==early]['value_mean']
    after=allpoints['norm_GFP_mem'][allpoints['norm_GFP_mem']['cond_bin']==late]['value_mean']
    
    tstat, pval = stats.ttest_rel(before, after)

    print(f"{early} t = {tstat:.3f}, p = {pval:.4g}")
```
