```python
import numpy as np
from scipy import stats

comps = ['231_17-5C','235_17-5C','235_17'] # change as needed

for cond in comps:
    sub = baseline[baseline['condition'] == cond]

    # drop incomplete pairs
    sub = sub.dropna(subset=['baseline_pre', 'baseline_post'])

    before = sub['baseline_pre']
    after  = sub['baseline_post']

    tstat, pval = stats.ttest_rel(before, after)
    mean1, mean2 = before.mean(), after.mean()

    print(
        f"{cond}: "
        f"mean={mean1:.3f} vs {mean2:.3f}, "
        f"t = {tstat:.3f}, p = {pval:.4g}"
    )
```
